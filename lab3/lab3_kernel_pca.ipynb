# KernelPCA

- Visualize in 2D datasets used in this lab, experiment with the parameters of the KernelPCA method change kernel and gamma params.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.decomposition import PCA

X, y = make_moons(n_samples=500, noise=0.02, random_state=417)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.title("PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(kernel="rbf", gamma=15)
X_kpca = kpca.fit_transform(X)

plt.title(f"Kernel PCA | kernel = {kpca.kernel}, gamma = {kpca.gamma}")
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y)
plt.show()
```

We can see different transformations done by different kernels and different values of gamma. <br>
Generally, the higher gamma, the more distorted is the data from the original. <br>
The more the effects of the kernel transformations are visible

```python
kernels = ["linear", "poly", "rbf", "cosine"]
gammas = [16, 8, 4, 2, 1, 1 / 2, 1 / 4]

for kernel in kernels:
    if kernel in ["poly", "rbf", "sigmoid"]:
        for gamma in gammas:
            kpca = KernelPCA(kernel=kernel, gamma=gamma)
            X_kpca = kpca.fit_transform(X)
            plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y)
            plt.title(f"Kernel PCA | kernel = {kpca.kernel}, gamma = {kpca.gamma}")
            plt.show()
    else:
        kpca = KernelPCA(kernel=kernel)
        X_kpca = kpca.fit_transform(X)
        plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y)
        plt.title(f"Kernel PCA | kernel = {kpca.kernel}")
        plt.show()
```
